## Homework Hack

1. Make a text analyzer with strings that accepts input from the user
- Takes multiple words/ sentences
- Displays original string
- Counts total characters (includes spaces)
- Displays longest word with how many characters it has
- Displays the string reversed
- Find middle word/ character (excludes spaces or special characters)
- Correctly displays everything!
2. Other criteria:
- make sure program can process upper/lowercase letters, numbers, and symbols
- Customize and add your own unique function
- Try to use both python and js, either making two separate analyzers or using both at the same time

```python
# Python Version
# Function to clean the string by removing non-alphanumeric characters (keeping spaces)
def clean_string(input_string):
    # Loop through each character and keep only alphanumeric characters and spaces
    cleaned = ''.join(c for c in input_string if c.isalnum() or c.isspace())
    return cleaned

# Function to find the middle word or character
def find_middle(input_string):
    clean_str = clean_string(input_string).split()

    if len(clean_str) == 0:  # Avoid empty string
        return "No middle word available."

    mid_index = len(clean_str) // 2
    if len(clean_str) % 2 == 0:
        # If even number of words, return the two middle words
        return f"Middle words: {clean_str[mid_index - 1]} and {clean_str[mid_index]}"
    else:
        # If odd number of words, return the middle word
        return f"Middle word: {clean_str[mid_index]}"

# Function to find the longest word in a string
def longest_word(input_string):
    words = clean_string(input_string).split()
    longest = max(words, key=len)  # Find the longest word
    return longest, len(longest)

# Main function to run the text analyzer
def text_analyzer():
    # Take input from the user
    user_input = input("Enter multiple words or sentences: ")

    # Display the original string
    print(f"\nOriginal String: {user_input}")

    # Count total characters including spaces
    total_characters = len(user_input)
    print(f"Total Characters (including spaces): {total_characters}")

    # Display the longest word and its length
    longest, longest_len = longest_word(user_input)
    print(f"Longest word: {longest} with {longest_len} characters")

    # Display the string reversed
    reversed_string = user_input[::-1]
    print(f"Reversed String: {reversed_string}")

    # Find and display the middle word/character
    middle = find_middle(user_input)
    print(middle)

    # Custom function: Count digits in the string
    digits_count = sum(c.isdigit() for c in user_input)
    print(f"Total digits in the string: {digits_count}")

# Run the text analyzer
text_analyzer()
```

```python
// JavaScript Version

// Function to clean the string by removing non-alphanumeric characters (keeping spaces)
function cleanString(inputString) {
    let cleaned = '';
    for (let i = 0; i < inputString.length; i++) {
        if (/[A-Za-z0-9\s]/.test(inputString[i])) {
            cleaned += inputString[i];
        }
    }
    return cleaned;
}

// Function to find the middle word or character
function findMiddle(inputString) {
    const cleanStr = cleanString(inputString).split(' ');
    
    if (cleanStr.length === 0) return "No middle word available.";
    
    const midIndex = Math.floor(cleanStr.length / 2);
    if (cleanStr.length % 2 === 0) {
        return `Middle words: ${cleanStr[midIndex - 1]} and ${cleanStr[midIndex]}`;
    } else {
        return `Middle word: ${cleanStr[midIndex]}`;
    }
}

// Function to find the longest word in a string
function longestWord(inputString) {
    const words = cleanString(inputString).split(' ');
    let longest = '';
    for (let i = 0; i < words.length; i++) {
        if (words[i].length > longest.length) {
            longest = words[i];
        }
    }
    return longest;
}

// Main function to run the text analyzer
function textAnalyzer() {
    // Take input from the user
    const userInput = prompt("Enter multiple words or sentences:");

    // Display the original string
    console.log(`\nOriginal String: ${userInput}`);

    // Count total characters including spaces
    const totalCharacters = userInput.length;
    console.log(`Total Characters (including spaces): ${totalCharacters}`);

    // Display the longest word and its length
    const longest = longestWord(userInput);
    console.log(`Longest word: ${longest} with ${longest.length} characters`);

    // Display the string reversed
    const reversedString = userInput.split('').reverse().join('');
    console.log(`Reversed String: ${reversedString}`);

    // Find and display the middle word/character
    const middle = findMiddle(userInput);
    console.log(middle);

    // Custom function: Count digits in the string
    const digitsCount = (userInput.match(/\d/g) || []).length;
    console.log(`Total digits in the string: ${digitsCount}`);
}

// Run the text analyzer
// textAnalyzer();  // Uncomment to run in browser environment
```
