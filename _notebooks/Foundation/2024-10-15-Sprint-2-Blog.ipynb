## Why Mr. Mortensens Self-Study Lesson is Relevant

This lesson is important as it teaches foundational concepts like MetaData, Classes, and For-Loops, which are essential for managing data and structuring code effectively. MetaData helps store and describe important information about resources, such as sprite images in game development. Classes allow you to group related properties and methods, making the code modular, reusable, and easy to maintain. For-Loops are used to iterate over data structures like arrays, enabling efficient processing of large datasets or grid-based elements like sprite sheets. These concepts work together to create interactive game elements, such as displaying animations from a sprite sheet. Mastering these techniques is useful not only in game development but also in web development, data science, and general software engineering. Understanding how to use these tools will help you write efficient, maintainable code for a wide range of applications.

Another challenge I faced was finding useful information about my lesson topic. After going to the College Board's "lessons" on Big Ideas 3.10 to see what I was supposed to teach, I was disappointed to find that there was very little helpful information. So I looked at other websites on how they taught Big Idea 3.10, such as Khan Academy. I then took information from these other websites and used them to inspire my own lesson, but taught it in a way that I understood and believed others would understand as well, even if they did not have previous coding experience.

Furthermore, I faced some challenges on my homework and popcorn hacks. One challenge was the amount of time I had to spend on these hacks. I approximately spent 4+ hours doing all of these homework and popcorn hacks due to the fact that most of the coding being taught was new to me. This caused me to continuously go back to that students lesson, and sometimes the lesson did not explain a concept in a way that I understood fully. Another challenge I faced during coding was when I had to create a choose-your-own adventure game that relied on user input and had multiple storylines. I struggled to find out how to get user input and then have multiple story lines, and I spent a little too much time making a story that I found interesting and fun.
